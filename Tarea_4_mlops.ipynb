## Definición

Para entender qué es MLOPS, necesitamos conocer el concepto de Devops primero. Devops se refiere a una combinación de prácticas, herramientas y estrategias para incrementar la capacidad de una organización para entregar aplicaciones y servicios a una mayor velocidad y de manera eficiente en vez de usar procesos de desarrollo de software tradicional.

Entonces, la idea de MLOPS es aplicar estos mismos principios de devops para poder entregar modelos de machine learning de una manera más eficiente y de mejor calidad. Involucra la colaboración y comunicación entre científicos de datos, ingenieros de Devops y de software. 

Aplicar estas prácticas incrementa la calidad, simplifica la administración y automatiza el despliegue de modelos de aprendizaje automático en ambientes productivos.

## Principios

* **Control de versiones del código, datos y resultados de los experimentos**: Adicional a versionar el código base, es importante realizar un versionamiento de los sets de datos, así como los modelos resultantes en los experimentos, esto puede ahorrar esfuerzo y costo computacional si se debe de recrear.
* **Utilizar múltiples ambientes**: Se debe segregar los ambientes de desarrollo, pruebas y producción, la infraestructura de producción debería de ser replicada en pruebas.
* **Manejar la infraestructura y configuraciones como código**: Utilizar IaC para prevenir inconsistencias entre ambientes. Las especificaciones para los experimentos deberían de manejarse como CaC de manera que se pueda re ejecutar y re-utilizar versiones de los experimentos en diferentes ambientes.
* **Administrar y rastrear los experimentos**: Se debe de mantener un rastreo de los KPIs y otros artefactos de los experimentos. Esto ayuda a tener un análisis cuantitativo de estos permitiendo una mejor colaboración y agilidad.
* **Probar el código, validar la integridad de los datos y la calidad del modelo**: Se debe de probar el código base que incluye las funciones para corrección de datos, extracción de características, validaciones de integridad de datos y obtención del rendimiento del modelo.
* **Integración y entrega continua (CI-CD)**: Utilizar CI para automatizar la ejecución de las pruebas. El entrenamiento del modelo se debe de incluir en un pipeline e incluir pruebas A/B como parte del despliegue.
* **Monitorear los servicios, modelos y datos**: Es crítico monitorear los servicios para ver tiempos en la infraestructura, así como para ver la calidad del modelo. Se deben de monitorear los datos para ver si empieza a haber una varianza que indique que se debe realizar un reentrenamiento o se pueden incluir triggers que ejecuten estos reentrenamientos.



## Etapas del ciclo de Aprendizaje Automatico que Involucra

* Obtención de datos
* Análisis de datos
* Transformación o preparación de los datos
* Desarrollo y entrenamiento del modelo
* Validación del modelo
* Despliegue del modelo
* Monitoreo del modelo
* Re-entrenamiento del modelo

## Beneficios

* **Eficiencia**: Desarrollo de modelos más veloz.
* **Escalabilidad**: Ayuda a escalar la cantidad de aplicaciones que utilizan aprendizaje automático.
* **Confianza**: Genera confianza al crear modelos en ambientes dinámicos al crear procesos repetitivos mediante automatizaciones, pruebas y validación.
* **Mejora la comunicación e integración**: Como DevOps, sigue patrones que buscan integrar el ciclo de desarrollo y de operaciones.
* **Mejor uso de los datos**: Reduce el ciclo de vida productivo generando información confiable que puede ser utilizada más rápidamente. También permite una mejor retroalimentación ayudando a determinar que puede ser ruido y que pueden ser anomalías que requieren atención.
* **Cumplimiento de politicas**: Permite reproducir modelos que cumplan con las regulaciones que sean necesarias mediante la creación de reglas en los pipelines para cumplir con dichas regulaciones.
* **Reducción de sesgo**: Permite aplicar políticas que protegen contra el sesgo de los datos y el modelo.


## Diferencias con respecto al ciclo de DevOps

Existen similitudes entre MLOps y DevOps dado que MLOps deriva de DevOps, sin embargo, existen diferencias entre ambos:

* **Habilidades del equipo**: En un proyecto de aprendizaje automático, los científicos de datos están enfocados en análisis de datos, desarrollo de modelos y experimentación por lo que puede que no tengan la experiencia de un ingeniero de software acostumbrado a generar servicios productivos
* **Desarrollo**: Por la naturaleza de aprendizaje automático, en el desarrollo del modelo se deben utilizar diferentes características, algoritmos, modelos, así como diferentes parámetros e hiper parámetros para determinar qué funciona mejor para el modelo. Debido a esto se debe mantener un registro de qué funcionó y que no, de manera que se pueda mantener la reusabilidad mientras se maximiza la reutilización de código.
* **Pruebas**: A diferencia de otros desarrollos de software, en aprendizaje automático, adicional a las pruebas unitarias y de integración, se necesita validación de datos, evaluar la calidad del modelo, y validar el modelo.
* **Despliegue**: El despliegue puede que requiera no solo el despliegue del modelo, sino que además pipelines que permitan re-entrenar y re-desplegar el modelo de manera automática.
* **Producción**: Debido a que los modelos pueden decaer de manera más rápida que el software tradicional, se deben de mantener estadísticas de los datos y monitorear el rendimiento del modelo que envíen notificaciones si el rendimiento no es el esperado.

*En lo que respecta a CI, CD, CT las diferencias son:*

* En CI, no solo se refiere a validar y probar componentes de código, sino que también se debe validar datos, esquemas y modelos.
* CD no solo representa un servicio de software, sí que se debe desplegar el sistema de entrenamiento que debe de desplegar de manera automática el modelo
* CT hace referencia a re-entrenar y servir los modelos de manera automática

## Niveles de implementación

### Nivel 0 - Manual

Hace referencia a un ambiente en donde el proceso de construcción y despliegue es manual

#### Sus características son: 
* **Manual, orientado a scripts, iterativo**: Cada paso es manual, esto incluye el análisis de los datos, la preparación de los datos, el entrenamiento y la validación. La transición de un paso al otro también es manual. Usualmente es código ejecutado en notebooks de manera interactiva por los científicos de datos.
* **No hay conexión entre ML y operaciones**: Separa a los científicos de datos que crean el modelo y los ingenieros que lo sirven como servicio. El científico entrega un modelo entrenado como un artefacto al equipo de ingenieros que lo despliegan en la infraestructura productiva.
* **Iteraciones de release poco frecuentes**: Se asume que son pocos modelos que no requieren cambios constantes, por lo que nuevas versiones del modelo se despliegan pocas veces durante el año.
* **Sin CI ni CD**: Debido a que se asumen pocos cambios, se omite el CI y el CD.
* **Despliegue hace referencia solo al modelo**: Solo se realiza un despliegue del modelo entrenado y no de todo el sistema de aprendizaje automático.
* **Poco monitoreo**: No se lleva registro de las predicciones o acciones del modelo requeridos para detectar degradación en este.

### Nivel 1 - Automatización del Pipeline de Apendizaje Automático

El objetivo es realizar entrenamiento continuo del modelo automatizando el pipeline de ML. Para automatizar este proceso, se debe introducir validación de datos y del modelo de manera automatizada.

#### Sus características son:

* **Rápida experimentación**: Los pasos para realizar los experimentos están orquestados. La transición entre estos está automatizado lo cual permite realizar iteraciones de los experimentos de manera rápida y acelera el paso a producción
* **CT del modelo**: El modelo es entrenado de manera automática en producción, utilizando datos frescos.
* **Simetría entre la experimentación y la operación**: El pipeline utilizado en el desarrollo o experimentación, es utilizado en pre producción y en producción.
* **Código modularizado**: Para construir los pipelines, los componentes deben de ser reutilizables y posiblemente compartidos entre pipelines, debido a esto, el EDA puede seguir realizándose en notebooks, pero el resto de código no. Idealmente los componentes deben de estar contenerizados.
* **CD del modelo**: Se entregan modelos entrenados con datos nuevos de manera constante como servicios de predicción.
* **Pipeline de despliegue**: Ya no se despliega el modelo como servicio de predicción únicamente, si no que se despliega todo un pipeline de entrenamiento que de manera automática y continua entrega un modelo como servicio de producción.

#### Los componentes requeridos son

##### *Validación del modelo y de los datos*
Debido a que se generan nuevas versiones de los modelos de manera continua, entrenados con datos recientes, el proceso de validación de datos y del modelo se debe de automatizar
La validación de datos debe decidir si el modelo debe ser re-entrenado o no y la decisión debe de tomarse de manera automática. Pueden suceder dos cosas, se detecte que los datos son anomalías por lo que el pipeline se detiene, o se detecta que hay cambios significativos en las propiedades estadísticas de los datos en cuyo caso se dispara un reentrenamiento.
La validación del modelo ocurre si el entrenamiento se realizó de manera satisfactoria, esto involucra generar métricas para medir la calidad del modelo, comparar estas métricas con el modelo actual (debe de tener un mejor rendimiento que el actual), asegurar que el nuevo modelo tenga compatibilidad con la infraestructura actual.

##### *Almacén de funciones*
Un almacén de funciones consiste en un repositorio centralizado donde se estandariza la definición, almacenamiento y acceso a las funciones de entrenamiento y uso.
El beneficio de este almacén es que permite descubrir y reusar funciones en vez de recrearlas o crear nuevas similares y evita diferencias en el entrenamiento y producción al utilizar las funciones en el almacén para experimentación, entrenamiento continuo y producción.

##### *Administración de la metadata*
Se debe almacenar información sobre cada ejecución del pipeline para facilitar la reusabilidad y las comparaciones. Esta información es:
* Versiones que fueron ejecutadas
* Fecha inicio y de fin de la ejecución del pipeline
* Quien ejecutó el pipeline
* Los parámetros que fueron pasados al pipeline
* Ubicación de datos, resultados en la validación, estadísticas generadas, vocabulario extraído de las características categóricas de manera que si el pipeline se detiene, se puede saber en qué punto fue y no re-ejecutar los pasos que fueron exitosos.
* Indicador de cual versión estaba previamente en caso de que se requiera un rollback
* Las métricas de evaluación del nuevo modelo

##### *Disparadores de las pipelines de ML*
Se puede automatizar las pipelines de producción para reentrenar modelos acorde a lo requerido:
* Bajo demanda
* Programado
* Cuando hay degradación del modelo
*Cuando se detecta que hay cambios significantes en la distribución de los datos

### Nivel 2 - Automatización del pipeline del CI y CD

Automatizar los pipelines de CI y CD permite que los científicos de datos exploren nuevas ideas y puedan implementarlas y automáticamente construirlas, probarlas y desplegar los nuevos componentes del pipeline al ambiente destino.

Las etapas de este nivel son:
* **Desarrollo y experimentación**: Se prueban nuevos algoritmos de ML y nuevos modelos donde los pasos experimentales son orquestados. La salida de esta etapa el código fuente de los pasos del pipeline, que luego son enviados al repositorio de código
* **Pipeline CI**: Se crea código fuente y se ejecutan múltiples pruebas, esto genera componentes del pipeline como paquetes, ejecutables y artefactos que serán desplegados.
* **Pipeline CD**: Se despliegan los insumos generados en la etapa previa. La salida es un pipeline con la nueva implementación del modelo.
Disparadores automatizados: El pipeline se ejecuta de manera automática en producción con base a un horario o en respuesta a un disparador. El modelo generado se envía al registro.
* **CD del modelo**: Se pone el modelo entrenado como servicio de producción. El resultado es el modelo desplegado.
* **Monitoreo**: Toma estadísticas del rendimiento del modelo en tiempo real. Esto puede disparar un pipeline o generar un nuevo ciclo de experimentos.
