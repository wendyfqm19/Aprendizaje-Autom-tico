- **SageMaker Studio.** Entorno de desarrollo totalmente integrado (IDE) para machine learning.

Amazon SageMaker Studio proporciona una interfaz visual única basada en la Web donde puede realizar todos los pasos de desarrollo de ML, incluida la creación de cuadernos, la administración de experimentos, la creación automática de modelos, la depuración y la detección de desviaciones de modelos y datos.

- **SageMaker Studio Lab.** Para aprender y experimentar con el aprendizaje automático. 

Amazon SageMaker Studio Lab es un entorno de desarrollo de ML gratuito que proporciona informática, almacenamiento (hasta 15 GB) y seguridad, todo sin costo alguno, para que cualquier persona pueda aprender y experimentar con ML. Lo único que necesita para comenzar es una dirección de email válida. No es necesario configurar infraestructura ni administrar la identidad y el acceso.

- **SageMaker Canvas.** Generar predicciones precisas de machine learning, sin necesidad de código. 

Amazon SageMaker Canvas proporciona una interfaz visual de apuntar y hacer clic que permite a los analistas empresariales generar predicciones de machine learning precisas por sí mismos sin necesidad de experiencia de ML ni de tener que escribir una sola línea de código.

- **SageMaker RStudio.** IDE de RStudio completamente administrado del sector en la nube para ciencia de datos y machine learning. 

RStudio es un entorno de desarrollo integrado para R, con una consola, editor de resaltado de sintaxis compatible con la ejecución de código directo y herramientas para el trazado, el historial, la depuración y la administración de espacios de trabajo. Amazon SageMaker admite RStudio como un entorno de desarrollo integrado (IDE) completamente administrado integrado con Amazon SageMaker Domain.
